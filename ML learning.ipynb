# 첫번째 예제

```python
# 필요한 모듈 임포트
import pandas as pd
from sklearn import svm, metrics
from sklearn.model_selection import train_test_split
```

```python
# 붓꽃 data 읽어 들이기
csv = pd.read_csv('https://raw.githubusercontent.com/rhiever/Data-Analysis-and-Machine-Learning-Projects/master/example-data-science-notebook/iris-data.csv')
csv = csv.dropna()
```

```python
# 필요한 열 추출하기
csv_data = csv[['sepal_length_cm','sepal_width_cm','petal_length_cm','petal_width_cm']]
csv_label = csv['class']
```

```python
# 학습전용 데이터와 테스트용 데이터 분리하기
train_data, test_data, train_label, test_label = train_test_split(csv_data,csv_label)
```

```python
# 데이터 학습시키고 예측하기
clf = svm.SVC()
clf.fit(train_data, train_label)
pre = clf.predict(test_data)

# 정답률 구하기
ac_score = metrics.accuracy_score(test_label,pre)
print('정확도 =', ac_score)
```

# 두번째 예제: 손글씨 숫자 인식


```python
import gzip, os, os.path
path = './samples/MNIST_data/'
files = [
    'train-images-idx3-ubyte.gz',
    'train-labels-idx1-ubyte.gz',
    't10k-images-idx3-ubyte.gz',
    't10k-labels-idx1-ubyte.gz'
    
]
# gzip 압축 해제
for f in files:
    gz_file = path + '/' + f
    raw_file = path + '/' + f.replace('.gz','')
    print('gzip: ', f)
    with gzip.open(gz_file,'rb') as fp:
        body = fp.read()
        with open(raw_file, 'wb') as w:
            w.write(body)
    print('unzip!')
```

```python
import struct

path = './samples/MNIST_data/'

def to_csv(name, maxdata):
	lbl_f = open(path+name+"-labels-idx1-ubyte", "rb")
	img_f = open(path+name+"-images-idx3-ubyte", "rb")
	csv_f = open(path+name+".csv", "w", encoding="utf-8")
	mag, lbl_count = struct.unpack(">II", lbl_f.read(8))
	mag, img_count = struct.unpack(">II", img_f.read(8))
	rows, cols = struct.unpack(">II", img_f.read(8))
	pixels = rows * cols

	res = []
	for idx in range(lbl_count):
		if idx > maxdata: break
		label = struct.unpack("B", lbl_f.read(1))[0]
		bdata = img_f.read(pixels)
		sdata = list(map(lambda n: str(n), bdata))
		csv_f.write(str(label)+",")
		csv_f.write(",".join(sdata)+"\r\n")

		if idx < 10:
			s = "P2 28 28 255\n"
			s += " ".join(sdata)
			iname = path+"{0}-{1}-{2}.pgm".format(name, idx, label)
			with open(iname, "w", encoding="utf-8") as f:
				f.write(s)
	csv_f.close()
	lbl_f.close()
	img_f.close()

to_csv("train", 10000)
to_csv("t10k", 500)
```

```python
# 이미지 데이터 학습시키기
from sklearn import model_selection, svm, metrics
path = './samples/MNIST_data/'
# csv 파일을 읽어 들이고 가공하기
def load_csv(fname):
	labels = []
	images = []
	with open(fname, "r") as f:
		for line in f:
			cols = line.split(",")
			if len(cols) < 2: continue
			labels.append(int(cols.pop(0)))
			vals = list(map(lambda n: int(n) / 256, cols))
			images.append(vals)
	return {"labels":labels, "images":images}

data = load_csv(path + "train.csv")
test = load_csv(path + "t10k.csv")

# 학습하기
clf = svm.SVC()
clf.fit(data["images"], data["labels"])

predict = clf.predict(test["images"])

ac_score = metrics.accuracy_score(test["labels"], predict)
cl_report = metrics.classification_report(test["labels"], predict)
print("정답률 = ", ac_score)
print("리포트 = ")
print(cl_report)    
```
