## 기하객체
`ggplot2`는 다양한 기하 객체들로 구성되어 있습니다. 
- `geom_area()`

```r
library(ggplot2)

df <- data.frame(
    x  = c(1,2,3),
    y  = c(2,4,6),
    lable = c('A', 'B', 'C'))

plt <-  ggplot(df, aes(x, y, label = label)) +
     labs( x = NULL, y = NULL) + # x축 레이블 숨기기
     theme(plot.title = element_text(size = 11)) # 제목 텍스트 사이즈

plt + geom_bar(stat = 'identity')
```

## 복잡성 추가

이번에는 약간더 복잡한 facet, 다중 레이어, 통계값을 혼합한 예를 살펴봅니다. 

```r
ggplot(mpg, aes(displ, hwy)) +
    geom_point() +
    geom_smooth(method = 'loess') + # 비선형 회귀 분석
    facet_wrap(~year)
```
